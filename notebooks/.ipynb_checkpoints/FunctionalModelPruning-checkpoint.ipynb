```python
def calculateNetDistances(data):
    new = {}
    graphs = {}
    adj_matrices = {}
    for index in data.index.values:
        new[index] = data.loc[index].unstack()
        #print(index)
        df2 = pd.concat([new[index], new[index].T], sort=True).fillna(0).sort_index() #ADDED a sort - energy harvesters rerun using tag "sorted" ... :(
        graphs[index] = nx.from_numpy_matrix(df2.values)
        adj_matrices[index] = nx.adjacency_matrix(graphs[index])
        mapping = dict(zip(graphs[index], df2.columns.values)) 
        graphs[index] = nx.relabel_nodes(graphs[index], mapping)

    ##Calculate the Laplacian Spectral Distance (pnorm 2) using NetComp 
    df = pd.DataFrame()
    for key1 in adj_matrices:
        for key2 in adj_matrices:
            A1 = adj_matrices[key1]
            A2 = adj_matrices[key2]
            d = nc.lambda_dist(A1,A2,kind='laplacian')
            df.loc[key1, key2] = d
            #print(key1, key2, d)
    #print(df) 

    ##Calculate the NetSimile Distance using NetComp
    netsimile = pd.DataFrame()
    for key1 in graphs:
        for key2 in graphs:
            G1 = graphs[key1]  
            G2 = graphs[key2]
            A1 = adj_matrices[key1]
            A2 = adj_matrices[key2]
            ns = nc.netsimile(A1,A2)
            netsimile.loc[key1, key2] = ns
            #print(key1, key2, ns)
    #print(netsimile)

    deltacon = pd.DataFrame()
    for key1 in graphs:
        for key2 in graphs:
            G1 = graphs[key1]  
            G2 = graphs[key2]
            A1 = adj_matrices[key1]
            A2 = adj_matrices[key2]
            dc = nc.deltacon0(A1,A2)
            deltacon.loc[key1, key2] = dc
            #print(key1, key2, dc)
    #print(deltacon) 
    return df, netsimile, deltacon
```

```python
def calculateVecDistances(data):
    hamming_sim_full = pd.DataFrame(squareform(pdist(data, metric='hamming')), columns = data.index, index = data.index)
    jaccard_sim_full = pd.DataFrame(squareform(pdist(data, metric='jaccard')), columns = data.index, index = data.index)
    cosine_sim_full = pd.DataFrame(squareform(pdist(data, metric='cosine')), columns = data.index, index = data.index)
    return hamming_sim_full, jaccard_sim_full, cosine_sim_full
```

```python
def rangeNorm(data):
    new = 1 - (data - data.min().min())/(data.max().max() - data.min().min())
    return new
```

```python
def hist(data, meas):
    new = data.where(np.triu(np.ones(data.shape),k=1).astype(np.bool))
    new_flat = new.to_numpy().flatten()
    final = new_flat[~np.isnan(new_flat)]
    sns.distplot(final,  hist=False, kde=True, kde_kws={'linewidth':4,"shade": True}, label=meas)
    return final
```

```python
def productrankingcorr(normsim1, normsim2, N):
    corrs = []
    sorted_1 = normsim1#.rank()
    sorted_2 = normsim2#.rank()
    for i in range(0,N):
        rankingcorr = sorted_1[sorted_1.columns[i]].corr(sorted_2[sorted_2.columns[i]], method='kendall')
        corrs.append(rankingcorr)
    n = len(corrs)
    reps = 500 #Number for bootstrapping
    np.random.seed(0)
    xb = np.random.choice(np.array(corrs), (n,reps)) #bootstrapping
    mb = xb.mean(axis=0)
    confint = np.percentile(mb, [2.5, 97.5])
    print(np.round(mb.mean(),2), np.round(confint,2))
    return  (corrs,np.round(confint,2),np.round(mb.mean(),2))
```

```python
a,b,c = calculateNetDistances(pruned_en)
d,e,f = calculateVecDistances(pruned_en)
labels = ['spectral','netsim','dc','smc','jac','cos']
dists = [a,b,c,d,e,f]
normed_dists = [rangeNorm(x) for x in dists]
for y in range(0,len(normed_dists)):
    hist(normed_dists[y],labels[y])
```

```python
labels = ['spectral','netsim','dc','smc','jac','cos']
for i in range(0,len(normed_dists)):
    for j in range(0,len(normed_dists)):
        print(labels[i],labels[j])
        productrankingcorr(normed_dists[i], normed_dists[j], 39)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import squareform


def plot_dendrogram(data, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    dists = squareform(data)
    linkage_matrix = linkage(dists, "single")

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)



# setting distance_threshold=0 ensures we compute the full tree.
# plot the top three levels of the dendrogram
plot_dendrogram(dists[5], labels=dists[0].index)
plt.xticks(rotation=90)
plt.show()
```
